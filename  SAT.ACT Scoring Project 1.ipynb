##### 27. Generate and interpret 95% confidence intervals for SAT and ACT participation rates.

```python
from scipy import stats
import scipy as sp
import numpy as np
import math
```

```python
s = merged_column_df['Participation(ACT)']
n, min_max, mean, var, skew, kurt = stats.describe(s)
std=math.sqrt(var)
confi_int_act = stats.norm.interval(0.05,loc=mean,scale=std)
```

```python
confi_int_act
```

```python
s = merged_column_df['Participation(SAT)']
n, min_max, mean, var, skew, kurt = stats.describe(s)
std=math.sqrt(var)
confi_int_sat = stats.norm.interval(0.05,loc=mean,scale=std)
```
